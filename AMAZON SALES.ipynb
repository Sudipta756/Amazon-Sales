# DATA CLEANING PROCESS

```python
# check unique values of "Unnamed: 22" column
print(df['Unnamed: 22'].unique())
```

```python
# drop unnecessary columns
df=df.drop(["index"],axis=1)
df=df.drop(["ship-postal-code"],axis=1)
df=df.drop(["Unnamed: 22"],axis=1)
df=df.drop(["fulfilled-by"],axis=1)
df=df.drop(["currency"],axis=1)
df=df.drop(["ASIN"],axis=1)
df
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
# drop rows in "Amount" column which contain null values
df=df[~np.isnan(df["Amount"])]
df
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
# drop rows which contain null values in categorical columns
df=df.dropna(subset=["ship-city","ship-state","ship-country","promotion-ids"])
df
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
df.info()
```

```python
# Force Date column to datetime safely
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
# Confirm conversion result
print(df['Date'].dtypes)
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df
```

```python
df.info()
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
# find outliers from Amount column
sns.boxplot(data=df,y=df["Amount"])
plt.title("FIND OUTLIERS FROM AMOUNT COLUMN")
plt.ylabel("AMOUNT")
plt.show()
```

```python
# detect outliers
q1=df["Amount"].quantile(0.25)
q3=df["Amount"].quantile(0.75)
iqr=q3-q1
print("q1:",q1)
print("q3:",q3)
print("iqr:",iqr)
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
print("lower_bound:",lower_bound)
print("upper_bound:",upper_bound)
```

```python
filter_outliers=df[(df["Amount"]>=lower_bound) & (df["Amount"]<=upper_bound)]
```

```python
sns.boxplot(y=filter_outliers["Amount"])
plt.title("REMOVE OUTLIERS FROM AMOUNT COLUMN")
plt.ylabel("AMOUNT")
plt.show()
```
